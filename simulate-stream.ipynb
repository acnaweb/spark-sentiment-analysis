### Simulando streaming TCP/IP 127.0.0.1:9898

```python
from pyspark.streaming import StreamingContext
```

```python
# intervalo batch de 1 segundo
batch_interval = 1
ssc = StreamingContext(sc, batch_interval)
```

#### Net Cat - Abre uma porta e permite input/digitacao na porta
#### Sintaxe: nc -lk [porta]

```python
# Criando socket DStream
# Receiver  - abre um socket
linhas = ssc.socketTextStream("127.0.0.1", 9898)
```

```python
type(linhas)
```

```python
palavras = linhas.flatMap(lambda line: line.split(" "))
```

```python
type(palavras)
```

```python
# conta cada palavra
pares = palavras.map(lambda palavra: (palavra, 1))
contaPalavras = pares.reduceByKey(lambda x, y: x + y)
```

```python
# imprime
contaPalavras.pprint()
```

```python
# inicia a coleta do stream de dados
ssc.start()

# aguarda a computacao ser finalizada
ssc.awaitTermination()
```
